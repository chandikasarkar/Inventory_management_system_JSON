```python
#importing Libraries
import json
import time

#Initializing variables
sales_details=(" ")

#loading Inventory records from json to string
file=open("Records.json",'r')
js=file.read()
file.close()

#converting string to Json
record=json.loads(js)

#Displaying MENU
print("------------------------MENU------------------------\n")
for key in record.keys():
    print(key,":",record[key]["Name"],"|", record[key]["Price"],"|",record[key]["Quantity"])
print("----------------------------------------------------\n")

#Taking user input about the purchase and User Details
ui_name=str(input("Enter Customer's Name"     ))
ui_mail=str(input("enter Customer's Mail Id"  ))
ui_ph  =str(input("Enter Customer's Ph Number"))
ui_pr  =str(input('Enter Product ID: '        ))
ui_qn  =int(input('Enter Quantity: '          ))

#if we have Enough Quantity
if(record[ui_pr]['Quantity'] >= ui_qn):
    
    print("--------------------------------------------------\n")
    print("                       BILL                       \n")

    print("Product Name     :", record[ui_pr]["Name"])
    print("Product Price(Rs):", record[ui_pr]["Price"],"Rs")
    print("Product Quantity :",ui_qn)
    print("----------------------------------------------------\n")
    print("Billing Amount(Rs):",  record[ui_pr]["Price"] * ui_qn,"Rs")
    print("----------------------------------------------------\n")

#Genareting Sales
    sales_details="Customer name: "+ui_name + ", Customer phone no: " + ui_ph + ", Customer mail id: " + ui_mail + ", Product ID: " + ui_pr + ", Product Quantity: " + str(ui_qn) + " total amount: ,"+str(record[ui_pr]["Price"] * ui_qn)+","+time.ctime()+"\n"

#Updating Records
    record[ui_pr]['Quantity']=record[ui_pr]['Quantity'] - ui_qn
    
#if we have Enough Quantity
else:

    print("sorry we are not having enough quantity")
    print("We are only having ",record[ui_pr]["Quantity"]," quantity")
    print("--------------------------------------------------\n")
    
    a=str(input("Press Y/y to purchase: "))
    if (a=="Y" or a=="y"):
    
        print("--------------------------------------------------\n")
        print("                       BILL                       \n")
    
        print("Name     :", record[ui_pr]["Name"])
        print("Price(Rs):", record[ui_pr]["Price"],"Rs")
        print("Quantity :",record[ui_pr]["Quantity"])
        print("----------------------------------------------------\n")
        print("Billing Amount(Rs):",  record[ui_pr]["Price"] * record[ui_pr]["Quantity"] ,"Rs")
        print("----------------------------------------------------\n")

#Genareting Sales
        sales_details="Customer name: "+ui_name + ", Customer phone no: " + ui_ph + ", Customer mail id: " + ui_mail + ", Product ID: " + ui_pr + ", Product Quantity: " + str(record[ui_pr]["Quantity"]) + " total amount: "+str(record[ui_pr]["Price"] * ui_qn)+","+time.ctime()+"\n"

#Updating Records    
        record[ui_pr]['Quantity']=0
        
    else:
        print("THANK YOU")
        
#Converting Dictionary to String
js=json.dumps(record)

#Saving Recods
file=open("Records.json",'w')
file.write(js)
file.close()

fd=open("Sales1.txt",'a')
fd.write(sales_details)
fd.close()

print("--------------------------------------------------\n")
print("THANKS FOR YOUR ORDER , INVENTORY UPDATED \n")
print("--------------------------------------------------\n")
```
